```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"]=(20,10)
```

```python
df1 = pd.read_csv("Bengaluru_House_Data.csv");
df1.head()
```

```python
df1.shape
```

```python
df1.groupby('area_type')['area_type'].agg('count')
```

```python
df2 = df1.drop(['area_type','society','balcony','availability'],axis='columns')
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3.isnull().sum()
```

```python
df3.shape
```

```python
df3['size'].unique()
```

```python
 df3['bhk']=df3['size'].apply(lambda x: int(x.split(' ')[0]))
```

```python
df3.head()
```

```python
df3['bhk'].unique()
```

```python
df3[df3.bhk>20]
```

```python
df3.total_sqft.unique()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df3[~df3['total_sqft'].apply(is_float)]
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens)==2:
        return(float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
convert_sqft_to_num('2166')
```

```python
convert_sqft_to_num('2100-2850')
```

```python
convert_sqft_to_num('6Acres')
```

```python
df4 = df3.copy()
df4['total_sqft']=df4['total_sqft'].apply(convert_sqft_to_num)
df4.head()
```

```python
df4.loc[30]
```

```python
df4.head(4)
```

```python
df5 = df4.copy()
df5['price_per_sqft'] = df5['price']*100000/df5['total_sqft']
df5.head()
```

```python
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x:x.strip())
location_stats = df5.groupby('location')['size'].agg('count').sort_values(ascending=False)
location_stats
```

```python
len(location_stats[location_stats<=10])
```

```python
location_stats_less_than_10=location_stats[location_stats<=10]
location_stats_less_than_10
```

```python
len(df5.location.unique())
```

```python
df5.location = df5.location.apply(lambda x: 'other' if x in location_stats_less_than_10 else x)
len(df5.location.unique())
```

```python
df5.head(10)
```

```python
df5.shape
```

```python
df6 = df5[~(df5.total_sqft/df5.bhk<300)]
df6.shape
```

```python
df6.price_per_sqft.describe()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m=np.mean(subdf.price_per_sqft)
        st=np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out

df7 = remove_pps_outliers(df6)
df7.shape
```

```python
def plot_scatter_chart(df,location):
    bhk2 = df[(df.location==location) &(df.bhk==2)]
    bhk3 = df[(df.location==location) & (df.bhk==3)]
    matplot.rcParams['figure.figsize']=(15,10)
    plt.scatter(bhk2.total_sqft,bhk2.price_per_sqft,color='blue',label='2 BHK',s=50)
    plt.scatter(bhk3.total_sqft,bhk3.price_per_sqft,marker='+',color='blue',label='3 BHK',s=50)
    plt.xlabel("Total Square Feet Area")
    plt.label("Price Per Square Feet")
    plt.title(location)
    plt.legen()
```
